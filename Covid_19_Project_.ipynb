```python
#Import Libraries 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Reading data 
df = pd.read_csv('/content/sample_data/Latest Covid-19 India Status.csv')
```

```python
df
```

```python
df.head()
```

```python
#Retrieving missing data 
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
df = df.sort_values(by=['Active'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Scatterplot with Active v/s Discharged 
plt.scatter(x=df['Active'], y=df['Discharged'])
plt.title('Covid Observation Chart')
plt.xlabel('Active Cases across Indian subcontinent')
plt.ylabel('Dicharged people safe from COVID')
plt.show()
```

```python
#Plot the same graph using Seaborn 
sns.regplot(x='Active', y='Discharged', data=df, scatter_kws={"color":"green"}, line_kws={"color":"blue"})
```

```python
#Determining correlation 
df.corr(method = 'pearson')
```
