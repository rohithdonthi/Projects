```python
#Using seaborn and matplotlib library 
import  seaborn as sns 
import matplotlib as plt
```

```python
sns.heatmap(data.isnull()) #which displays the null values in "state" column
```

```python
#Showing the number of confirmed, Deaths and Recovered cases of each Region 
data.groupby('Region').sum()
```

```python
data.groupby('Region')['Confirmed'].sum()
```

```python
#sorting the above data wrt the numbers in highest order 
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
```

```python
#TOP 10 countries of confimed cases
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False).head(10)
```

```python
data.groupby('Region')['Confirmed','Recovered'].sum()
```

```python
#Remove all the records where Confirmed cases are less than 10 
data = data[~(data.Confirmed <10)]
```

```python
data #displaying the data 
```

```python
#Region having maximum confirmed cases 
data.groupby('Region').Confirmed.sum().sort_values(ascending=False) #US has the highest number of cases 
```

```python
#Region having minimum number of cases 
data.groupby('Region').Deaths.sum().sort_values(ascending=True) #Laos has the minimum number of cases 
```

```python
#Sorting the confirmed cases of top 50 places in ascending order 
data.sort_values(by=['Confirmed'], ascending = True).head(50)
```

```python
#Recovery cases of top 50 places in Descending order 
data.sort_values(by=['Recovered'], ascending = False).head(50)
```

```python
#Importing libraries for chart representation 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Scatterplot for compariosions of Confirmed and Recovered cases
plt.scatter(x=data['Confirmed'], y=data['Recovered'])
plt.title('Covid Observation Chart')
plt.xlabel('Confirmed Cases across the World')
plt.ylabel('Patients recovered from COVID')
plt.show()
```

```python
#Plotting graph using Seaborn 
sns.regplot(x='Confirmed', y='Recovered', data=data, scatter_kws={"color":"Purple"}, line_kws={"color":"black"})
```
